Filter dan Replace serta konversi tipe data Object ke Float untuk IPS dan IPK

```python
# 1. Filter rows containing commas in 'IPS' column:
comma_rows = komputerfix[komputerfix['IPS'].astype(str).str.contains(',')]

# 2. Replace commas with periods in filtered rows (optional):
komputerfix.loc[comma_rows.index, 'IPS'] = komputerfix.loc[comma_rows.index, 'IPS'].str.replace(',', '.')

# 3. Convert 'IPS' column to float (if desired):
komputerfix['IPS'] = komputerfix['IPS'].astype(float)
#komputerfix['IPK'] = komputerfix['IPK'].astype(float)
```

```python
# 1. Filter rows containing commas in 'IPK' column:
comma_rows = komputerfix[komputerfix['IPK'].astype(str).str.contains(',')]
space_rows = komputerfix[komputerfix['IPK'].astype(str).str.contains(' ')]


# 2. Replace commas with periods in filtered rows (optional):
komputerfix.loc[comma_rows.index, 'IPK'] = komputerfix.loc[comma_rows.index, 'IPK'].str.replace(',', '.')
komputerfix.loc[space_rows.index, 'IPK'] = komputerfix.loc[space_rows.index, 'IPK'].str.replace(' ', '')

# 3. Convert 'IPK' column to float (if desired):
komputerfix['IPK'] = komputerfix['IPK'].astype(float)
```

```python
nbr_c = komputerfix.groupby(['PS', 'Semester', 'NPM','ST','Label']).agg(\
    Nbr=('Kode', 'count')).reset_index()\
    .pivot(index=['PS', 'Semester',  'NPM', 'ST'], columns='Label')\
    .fillna(0).droplevel(0, axis=1).reset_index()

nbr_c['EN'] = nbr_c['FA'] + nbr_c['NX'] + nbr_c['PA']

nbr_c.sample(n=5, random_state=0).style.set_table_styles()\
    .format('{:}', subset=['Semester', 'NPM', 'FA', 'NX', 'PA','EN'])\
    .hide(axis="index")
```

```python
nrc = nbr_c[['PS','NPM', 'Semester', 'ST', 'PA', 'EN']].rename(columns={'PA':'Passed', 'EN':'Enrolled'})

gr = komputerfix[komputerfix['Label'] != 'NX'].sort_values(['PS', 'ST'])

gr = gr.groupby(['PS', 'Semester', 'ST','NPM']).agg(Grade=('HM', 'median')).reset_index()
gr.columns = ['PS', 'Semester', 'ST', 'NPM', 'Median Grade']


data = pd.merge(nrc, gr, how='left')

data = data.set_index(['PS', 'Semester','ST', 'NPM']).stack().reset_index()\
    .rename(columns={'level_4': 'Agg', 0:'Value'})\
    .sort_values(['PS', 'Semester', 'ST', 'Agg'])

data['ST'] = data['ST'].map({'D':'Dropouts', 'G':'Graduates'})
data.sample(n=5, random_state=0).style.set_table_styles()\
    .format('{:}', subset=['Value'])\
    .hide(axis="index")
```
